# Aufgabe 3_3_1 ‚Äì Reaktion von PID und PT1 auf Einheitssprung

## Ziel
Beobachten Sie die Reaktion einer **PT1-Strecke** in offenem Regelkreis auf verschiedene Regler-Einstellungen.

## Modell-Aufbau (offener Regelkreis ‚Äì kein Feedback)

```
Step ‚îÄ‚îÄ‚ñ∂ PID(Kp, Ki, Kd) ‚îÄ‚îÄ‚ñ∂ PT1(œÑ=5) ‚îÄ‚îÄ‚ñ∂ Scope
```

| Block | Klasse | Parameter |
|-------|--------|-----------|
| Einheitssprung | `Step` | `t_step=1`, `final=1` |
| Regler | `PID` | `Kp=1`, `Ki=0`, `Kd=0` |
| Strecke | `PT1` | `tau=5` |

> Die PT1-√úbertragungsfunktion im Frequenzbereich: $G(s) = \frac{1}{\tau s + 1}$  
> Bei $t = \tau$ sind 63 % des Endwertes erreicht.

````python
import sys
sys.path.insert(0, '..')
from blockdiagram import Step, PID, PT1, Scope
````

````python
u       = Step(t_step=1, final=1)
regler  = PID(Kp=1, Ki=0, Kd=0, source=u)   # reiner P-Regler (Kp=1 ‚Üí kein Effekt)
strecke = PT1(tau=5, K=1, source=regler)

Scope(t_end=30, ylabel="Amplitude").run(
    Eingang=u,
    Ausgang=strecke
)
````

## Aufgaben

**a)** Variieren Sie `Kp` und `tau`: Wie beeinflusst die Zeitkonstante die Anstiegsgeschwindigkeit?

*Antwort:*

**b)** Schalten Sie den I-Anteil zu (`Ki > 0`): Was ver√§ndert sich?

*Antwort:*

````python
# Parameterstudie tau
for tau in [1, 5, 15]:
    u       = Step(t_step=1, final=1)
    regler  = PID(Kp=1, Ki=0, Kd=0, source=u)
    strecke = PT1(tau=tau, K=1, source=regler)
    Scope(t_end=30, title=f"œÑ = {tau} s").run(Eingang=u, Ausgang=strecke)
````

````python
# Mit I-Anteil
u       = Step(t_step=1, final=1)
regler  = PID(Kp=1, Ki=0.2, Kd=0, source=u)
strecke = PT1(tau=5, K=1, source=regler)
Scope(t_end=30, title="Mit I-Anteil (Ki=0.2)").run(Eingang=u, Ausgang=strecke)
````
