# Aufgabe 3_2_3 – Wassertank mit PID-Regler

## Ziel
Erweitern Sie das Wassertank-Modell (offene Steuerkette) zu einem **geschlossenen Regelkreis** mit PID-Regler.

## Erinnerung: Wassertank ohne Regelung

Ein 100 l fassender Tank ist zu Beginn mit 10 l gefüllt. Ein Zufluss integriert sich zum Füllstand auf.

## Modell-Aufbau (mit Regelkreis)

```
Step(w=60) ──▶ [Sum] ──▶ PID(Kp, Ki, Kd) ──▶ Integrator ──▶ Füllstand
                ↑                                                 │
                └───────────── Rückführung ───────────────────────┘
```

| Block | Klasse | Parameter |
|-------|--------|-----------|
| Sollwert | `Step` | `final=60` (gewünschter Füllstand in Liter) |
| Summationsstelle | `Sum` | $e = w - \text{Füllstand}$ |
| PID-Regler | `PID` | `Kp=1`, `Ki=0`, `Kd=0` |
| Strecke | `Integrator` | `initial=10` (Anfangsfüllstand) |
| Rückführung | direkter Pfad | — |

````python
import sys
sys.path.insert(0, '..')
from blockdiagram import Step, Sum, PID, Integrator, Scope
````

````python
# Regelkreis aufbauen
Kp = 1.0
Ki = 0.0   # zunächst 0 (reiner P-Regler)
Kd = 0.0

w         = Step(final=60)              # Sollwert: 60 Liter
e         = Sum()                       # Regelabweichung
regler    = PID(Kp=Kp, Ki=Ki, Kd=Kd, source=e)
fuellstand = Integrator(K=1, initial=10, source=regler)  # Tank, startet bei 10 l

e.connect(w,          sign=+1)
e.connect(fuellstand, sign=-1)   # Rückführung

Scope(t_end=30, ylabel="Füllstand [l]",
      title=f"Wassertank: Kp={Kp}, Ki={Ki}, Kd={Kd}").run(
    Sollwert=w,
    Fuellstand=fuellstand
)
````

````python
# Mit I-Anteil: Ki > 0
Kp = 1.0
Ki = 0.5   # ← anpassen
Kd = 0.0

w          = Step(final=60)
e          = Sum()
regler     = PID(Kp=Kp, Ki=Ki, Kd=Kd, source=e)
fuellstand = Integrator(K=1, initial=10, source=regler)
e.connect(w, +1)
e.connect(fuellstand, -1)

Scope(t_end=30, ylabel="Füllstand [l]",
      title=f"Wassertank: Kp={Kp}, Ki={Ki}, Kd={Kd}").run(
    Sollwert=w,
    Fuellstand=fuellstand
)
````
