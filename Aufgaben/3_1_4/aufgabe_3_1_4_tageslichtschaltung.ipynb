# Aufgabe 3_1_4 – Tageslichtschaltung als Einpunktregler

## Ziel
Übersetzen Sie die Tageslichtschaltung aus Aufgabe 2_3_1 in ein Blockschaltbild.
Als Schaltelement verwenden wir zunächst den einfachen `SinglePointController`.

## Modell-Aufbau

```
Sine(Außenhelligkeit) ▶ SinglePointController(threshold) ▶ Scope
```

| Block | Klasse | Parameter |
|-------|--------|-----------|
| Außenhelligkeit | `Sine` | Amplitude 250 Lux, Offset 250 Lux, Periode 24 h |
| Schwellwertschalter | `SinglePointController` | `threshold=220` |
| Zeitverlauf-Plot | `Scope` | `t_end=48` (zwei Tage) |

````python
import sys; sys.path.insert(0, "..")
from blockdiagram import Sine, SinglePointController, Scope
````

````python
# Außenhelligkeit: Sinus zwischen 0 und 500 Lux, Periode 24 Stunden
h_aussen = Sine(amplitude=250, offset=250, period=24)

# Schwellwertschalter: LED an wenn Helligkeit < 220 Lux
led = SinglePointController(threshold=220, source=h_aussen)

Scope(t_end=48, xlabel="Zeit [h]").run(
    Helligkeit_Lux=h_aussen,
    LED_an=led
)
````
