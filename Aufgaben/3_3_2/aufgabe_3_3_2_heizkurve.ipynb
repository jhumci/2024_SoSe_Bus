# Aufgabe 3_3_2 – Heizkurve (Steuerkette ohne Regelung)

## Situation

In einem ausgekühlten Gebäude wird die Raumtemperatur durch eine **Heizkurve** gesteuert  
(**keine Rückkopplung** – offene Steuerkette).

Die Heizkurve berechnet die Vorlauftemperatur aus der Außentemperatur:
$$T_{\text{VL}} = a \cdot T_{\text{außen}} + b \quad \text{mit } a = -1{,}5,\ b = 60$$

Das Stellglied (Heizanlage) reagiert träge: **PT1 mit τ = 60 min**.  
Der Raum selbst ist ebenfalls träge: **PT1 mit τ = 120 min, K = 0,35**.

## Modell-Aufbau

```
Step(Außentemp) ──▶ Gain(a=-1.5) ──▶ (+b=60) ──▶ PT1(τ=60) ──▶ PT1(τ=120, K=0.35) ──▶ Raumtemperatur
```

> **Hinweis:** `+b=60` = Addition einer Konstante → `Sum` mit einem `Constant`-Block.

````python
import sys
sys.path.insert(0, '..')
from blockdiagram import Step, Gain, Constant, Sum, PT1, Scope
````

````python
# Außentemperatur: fällt nach 1000 min von 10°C auf -5°C
T_aussen = Step(t_step=1000, initial=10, final=-5)

# Heizkurve: T_VL = -1.5 * T_aussen + 60
heizkurve = Sum(
    Gain(K=-1.5, source=T_aussen),
    Constant(60),
    signs=[+1, +1]
)

# Stellglied: träge Heizanlage
stellglied = PT1(tau=60, K=1, source=heizkurve)

# Strecke: Raumdynamik
raum = PT1(tau=120, K=0.35, source=stellglied, initial=10*0.35)

Scope(t_end=3000, ylabel="Temperatur [°C]", xlabel="Zeit [min]",
      title="Heizkurvensteuerung (offen – keine Rückkopplung)").run(
    Aussentemperatur=T_aussen,
    Vorlauftemperatur=stellglied,
    Raumtemperatur=raum
)
````
