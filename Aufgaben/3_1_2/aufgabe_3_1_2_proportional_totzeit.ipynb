## Ziel
Modellieren Sie eine **Reihenschaltung** aus Proportionalglied und Totzeitglied und untersuchen Sie den Einfluss der Parameter auf das Ausgangssignal.

## Modell-Aufbau

| # | Block | Klasse | Parameter |
|---|-------|--------|-----------|
| 1 | Sprungfunktion | `Step` | `t_step=1`, `final=1` |
| 2 | Proportionalglied | `Gain` | `K=2` |
| 3 | Totzeitglied | `TransportDelay` | `Tt=3` |
| 4 | Zeitverlauf-Plot | `Scope` | `t_end=15` |

Verbinden Sie die Blöcke in Reihe:
```
Step ──▶ Gain(K=2) ──▶ TransportDelay(Tt=3) ──▶ Scope
```

Stellen Sie Eingang **und** Ausgang im selben Diagramm dar.

````python
import sys
sys.path.insert(0, '..')   # blockdiagram.py liegt im Aufgaben-Ordner
from blockdiagram import Step, Gain, TransportDelay, Scope
````

````python
# Blöcke definieren und verknüpfen
u = Step(t_step=1, final=1)
gain = Gain(K=2, source=u)
y = TransportDelay(Tt=3, source=gain)

# Simulation und Darstellung
Scope(t_end=15, ylabel="Amplitude").run(
    Eingang=u,
    Ausgang=y
)
````
